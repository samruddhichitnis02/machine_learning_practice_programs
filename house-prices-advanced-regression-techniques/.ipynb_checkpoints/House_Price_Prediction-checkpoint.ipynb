```python
import seaborn as sb
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('train.csv')
```

```python
data.head()
```

```python
len(data['Id'].unique())
```

```python
data.drop(['Id'], axis = 1, inplace = True)
```

```python
data.shape
```

```python
fig, ax = plt.subplots(figsize = (10,8))
sb.heatmap(data.isna(),
          cbar = None,
          yticklabels = False,
          cmap = 'Blues')
plt.show()
```

# Droping the columns who has data  less than 50%

```python
data = data[[column for column in data if data[column].count() / len(data) >= 0.5]]
```

```python
data.shape
```

```python
fig, ax = plt.subplots(figsize = (10,8))
sb.heatmap(data.isna(),
          cbar = None,
          yticklabels = False,
          cmap = 'Blues')
plt.show()
```

```python
for i in data.columns:
    print(i)
```

# Getting the columns having Missing values along with their missing values

```python
Null_Values = [ ]
types = [ ]
for i in data.columns:
    if (data[i].isnull().sum() > 0):
        Null_Values.append(data[i].isnull().sum())
        types.append(data[i].dtype)
```

```python
pd.DataFrame({'Column': data.columns[data.isnull().any()],
              'Null_Values': Null_Values,
             'Datatype': types})
```

# Data Preprocessing

### Checking if their are outliers to handle missing values

```python
plt.figure(figsize=(20,8))
sb.boxplot(x = "LotFrontage", y = "SalePrice", data = data[data['LotFrontage'] < 90])
plt.show()
```

```python
plt.figure(figsize=(20,8))
sb.boxplot(x = "LotFrontage", y = "SalePrice", data = data[data['LotFrontage'] >= 90])
plt.show()
```

```python
data['LotFrontage'] = data['LotFrontage'].fillna(data['LotFrontage'].mean())
```

```python
plt.figure(figsize=(20,8))
sb.boxplot(x = "MasVnrArea", y = "SalePrice", data = data)
plt.show()
```

```python
data['MasVnrArea'] = data['MasVnrArea'].fillna(data['MasVnrArea'].mean())
```

```python
data['BsmtQual'] = data['BsmtQual'].fillna(data['BsmtQual'].mode()[0])
data['BsmtCond'] = data['BsmtCond'].fillna(data['BsmtCond'].mode()[0])
data['BsmtExposure'] = data['BsmtExposure'].fillna(data['BsmtExposure'].mode()[0])

data['MasVnrType'] = data['MasVnrType'].fillna(data['MasVnrType'].mode()[0])
data['BsmtFinType1'] = data['BsmtFinType1'].fillna(data['BsmtFinType1'].mode()[0])
data['BsmtFinType2'] = data['BsmtFinType2'].fillna(data['BsmtFinType2'].mode()[0])
```

```python
data['FireplaceQu'] = data['FireplaceQu'].fillna(data['FireplaceQu'].mode()[0])
data['GarageType'] = data['GarageType'].fillna(data['GarageType'].mode()[0])

data['GarageFinish'] = data['GarageFinish'].fillna(data['GarageFinish'].mode()[0])
data['GarageQual'] = data['GarageQual'].fillna(data['GarageQual'].mode()[0])
data['GarageCond'] = data['GarageCond'].fillna(data['GarageCond'].mode()[0])
```

```python
plt.figure(figsize=(20,8))
sb.boxplot(x = "GarageYrBlt", y = "SalePrice", data = data[data['GarageYrBlt'] < 1975])
plt.show()
```

```python
plt.figure(figsize=(20,8))
sb.boxplot(x = "GarageYrBlt", y = "SalePrice", data = data[data['GarageYrBlt'] >= 1975])
plt.show()
```

```python
data['GarageYrBlt'] = data['GarageYrBlt'].fillna(data['GarageYrBlt'].mean())
```

```python
fig, ax = plt.subplots(figsize = (10,8))
sb.heatmap(data.isna(),
          cbar = None,
          yticklabels = False,
          cmap = 'viridis')
plt.show()
```
